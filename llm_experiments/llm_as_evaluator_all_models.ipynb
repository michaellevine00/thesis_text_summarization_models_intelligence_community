# Llama3 as Expert Evaluator
Comparison of all models: BART, BART fine-tuned, T5, T5 fine-tuned, Pegasus, Llama2

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#instantiate a dataframe for the "all_models" analysis (df_all_models)
df_all_models=pd.read_csv('data/llm_evaluator_full_comparison_checkpoint.csv')
df_all_models.head(5)
```

```python
df_all_models.columns
```

```python
cols_to_compare = ['full_comparison_order_1_results', 'full_comparison_order_2_results',
       'full_comparison_order_3_results', 'full_comparison_order_4_results',
       'full_comparison_order_5_results', 'full_comparison_order_6_results']
```

```python
row_modes = df_all_models[cols_to_compare].mode(axis=1)
row_modes.tail(20)
```

```python
df_all_models.tail()
```

```python
df_all_models = pd.concat([df_all_models,row_modes], axis=1)

df_all_models
```

```python
df_all_models.rename(columns={0: 'mode_0', 1: 'mode_1'}, inplace=True)
df_all_models
```

```python
df_all_models['winner'] = np.where(df_all_models['mode_1'].isna(), df_all_models['mode_0'], 'none')
df_all_models
```

```python
(df_all_models
 ['winner']
 .value_counts()
)
```

```python
#Build new dataframe to show results of voting
data = {'model': ['llama_2', 'no_winner', 'pegasus', 'bart', 'bart_fine_tuned', 't5','t5_fine_tuned'],
        'value': [196,2,1,1,0,0,0]}

df_all_models_results = pd.DataFrame(data)
df_all_models_results
```

```python
fig = plt.figure()

fig.patch.set_facecolor('white')
fig.patch.set_alpha(0)
fig = plt.figure(figsize=(9,4),linewidth=10, facecolor='white')

plt.rcParams['figure.figsize'] = (8,4)
plt.rcParams['figure.dpi'] = 200
plt.rcParams
plt.style.context('seaborn-paper')

colors = sns.color_palette(['#B4B4B4'])
sns.barplot(data=df_all_models_results,y=df_all_models_results['model'],x=df_all_models_results['value'], palette = colors,width=0.4, dodge=False)
plt.xlabel('Number of times the model was chosen as best by LLM (n=200)')
plt.ylabel('')
plt.tick_params(top=False, labeltop=False, bottom=True, labelbottom=True)
#plt.tight_layout()
plt.legend('', frameon=False)
plt.suptitle('Llama3 LLM as Evaluator')
plt.title('Comparison of all models', fontsize=10)
#plt.grid(visible=True, axis='x', which='major' )
plt.annotate(text='196', xy=([400,400]), xycoords='axes points', xytext=(452,201), size=12, color='#FF6F27')
#plt.annotate(text='63', xy=([500,500]), xycoords='axes points', xytext=(290,106), size=15)
#plt.annotate(text='42', xy=([500,500]), xycoords='axes points', xytext=(189,31), size=15)
#plt.annotate(text='18', xy=([500,500]), xycoords='axes points', xytext=(60,31), size=15)
plt.annotate(text='For 2 records, LLM votes did not identify \na winner.  Models tied. ', xy=([20,175]), xycoords='axes points', xytext=(60, 165),
            arrowprops=dict(arrowstyle='->,widthB=6.2',linewidth=1, connectionstyle='bar,fraction=0'));# textcoords=('axes fraction'), ,
#            arrowprops=dict(arrowstyle='-[,widthB=6.2',color='#FF6F27',linewidth=1, connectionstyle='bar,fraction=0'))
plt.annotate(text='Llama3 preferred Llama2 \nsummaries in 196 of 200 records (98%) ', xy=([460,196]), xycoords='axes points', xytext=(285, 145), color='#FF6F27',
            arrowprops=dict(arrowstyle='->,widthB=6.2',linewidth=1, color='#FF6F27', connectionstyle='bar,fraction=0'));# textcoords=('axes fraction'), ,
plt.annotate(text='2', xy=([6,175]), xycoords='axes points', xytext=(6, 171))
            # arrowprops=dict(arrowstyle='-[,widthB=1.2',linewidth=1, connectionstyle='bar,fraction=0'))
plt.annotate(text='1', xy=([12,126]), xycoords='axes points', xytext=(30, 123),
            arrowprops=dict(arrowstyle='-[,widthB=2.2',linewidth=1, connectionstyle='bar,fraction=0'))
plt.annotate(text='0', xy=([12,43]), xycoords='axes points', xytext=(30, 40),
            arrowprops=dict(arrowstyle='-[,widthB=3.6',linewidth=1, connectionstyle='bar,fraction=0'))
plt.savefig('data/llm_eval_all_models.png', bbox_inches='tight');
```
