## Postprocessing

```python
import re
```

```python
def extract_number(text): 
    pattern = r'\((\d+)\)'

    matches = re.findall(pattern, text)

    if matches:
        return int(matches[0])
    else:
        return None
```

```python
cleaned_resp = extract_number(resp)
cleaned_resp
```

```python
def call_llm(prompt):
    
    try:
        base_url = 'http://opal.cmz.jhuapl.edu/'
        input_params = {"model": "meta-llama/Llama-2-70b-chat-hf", "prompt": prompt}
    
        with httpx.Client(base_url = url) as client:
            r= client.post(
                '/completions',
                json=input_params,
                timeout=None,
            )
            response=r.json()
        resp=response['choices'][0]['text']
        return resp
    except:
        return ""
```

```python
#feeds in three variables returns su
def build_comparison_3_prompt(article, summary_1, summary_2, summary_3):
    comparison_prompt = f"""You are an expert evaluator for judging summaries.  You will be given an article and three different summaries.  Your task is to choose which summary is the best.
Article:
{article}

Summary 1:
{summary_1}

Summary 2:
{summary_2}

Summary 3:
{summary_3}

Please respond with the number id surrounded by parenthesis of the summary that best summarized the original article. The best summary is summary number"""
    return comparison_prompt

    
```

```python
def evaluate_3_summaries(article, summary_1, summary_2, summary_3):
    #build prompt
    prompt = build_comparison_3_prompt(article, summary_1, summary_2, summary_3)

    #call llm w/ that prompt
    raw_resp = call_llm(prompt)

    #post-process the response
    cleaned_resp = extract_number(resp)

    return cleaned_resp
```

## Build loop to go through dataframe

```python
#define custom operation for pandas df
def evaluate_3_summaries_row_order1(row):
    article=row['article']
    summary_1 =row['t5_summaries']
    summary_2 =row['bart_summaries']
    summary_3 = row['pegasus_summaries']

    result = evaluate_3_summaries(article, summary_1, summary_2, summary_3)
    if result == 1:
        return "t5"
    elif result == 2:
        return "bart"
    elif result == 3:
        return "pegasus"
    else:
        return None

#define custom operation for pandas df
def evaluate_3_summaries_row_order2(row):
    article=row['article']
    summary_1 =row['t5_summaries']
    summary_3 =row['bart_summaries']
    summary_2 = row['pegasus_summaries']

    result = evaluate_3_summaries(article, summary_1, summary_2, summary_3)
    if result == 1:
        return "t5"
    elif result == 3:
        return "bart"
    elif result == 2:
        return "pegasus"
    else:
        return None

#define custom operation for pandas df
def evaluate_3_summaries_row_order3(row):
    article=row['article']
    summary_3 =row['t5_summaries']
    summary_2 =row['bart_summaries']
    summary_1 = row['pegasus_summaries']

    result = evaluate_3_summaries(article, summary_1, summary_2, summary_3)
    if result == 3:
        return "t5"
    elif result == 2:
        return "bart"
    elif result == 1:
        return "pegasus"
    else:
        return None

def evaluate_3_summaries_row_order4(row):
    article=row['article']
    summary_2 =row['t5_summaries']
    summary_1 =row['bart_summaries']
    summary_3 = row['pegasus_summaries']

    result = evaluate_3_summaries(article, summary_1, summary_2, summary_3)
    if result == 2:
        return "t5"
    elif result == 1:
        return "bart"
    elif result == 3:
        return "pegasus"
    else:
        return None

def evaluate_3_summaries_row_order5(row):
    article=row['article']
    summary_2 =row['t5_summaries']
    summary_3 =row['bart_summaries']
    summary_1 = row['pegasus_summaries']

    result = evaluate_3_summaries(article, summary_1, summary_2, summary_3)
    if result == 2:
        return "t5"
    elif result == 3:
        return "bart"
    elif result == 1:
        return "pegasus"
    else:
        return None

def evaluate_3_summaries_row_order6(row):
    article=row['article']
    summary_3 =row['t5_summaries']
    summary_1 =row['bart_summaries']
    summary_2 = row['pegasus_summaries']

    result = evaluate_3_summaries(article, summary_1, summary_2, summary_3)
    if result == 3:
        return "t5"
    elif result == 1:
        return "bart"
    elif result == 2:
        return "pegasus"
    else:
        return None
```

* Max context window of llama2-70b

* We can get around it by prompting in a certain way. so we pretend we are the AI model, and start it's response for it.
* difficult to constrain the response of an LLM.  It's giving us a number, but not giving us a number every time.
