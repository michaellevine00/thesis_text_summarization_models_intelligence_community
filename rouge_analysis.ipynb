```python

g = sns.FacetGrid(slope_df, col="model", col_wrap=4, height=4,) # ylim=(0, 10))
g.map(sns.pointplot, "state", "value", 'metric',color=".3", errorbar=None)
#g.fig.subplots_adjust(top=0.9)
plt.subplots_adjust(top=0.8)
g.fig.suptitle('TITLE!', y=1)
plt.tight_layout()
```

```python

g = sns.FacetGrid(slope_df, col="model", height=4,) # ylim=(0, 10))
g.map(sns.pointplot, 'state', "value", 'metric',color=".3", errorbar=None)
#g.fig.subplots_adjust(top=0.9)
plt.subplots_adjust(top=0.8)
g.fig.suptitle('ROUGE scores before and after fine-tuning', y=1)
g.tight_layout()
g.set_axis_labels("Before and after fine-tuning", "ROUGE scores")
```
