Comparing t5 vs t5 fine-tuned models

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#instantiate a dataframe for the "t5 vs. t5 fine_tuned" analysis (df_t5)
df_t5=pd.read_csv('data/llm_evaluator_t5_vs_t5_fine_tune_checkpoint.csv')
df_t5.head(5)
```

```python
df_t5.columns
```

```python
cols_to_compare = ['t5_vs_finetune_order_1_results', 't5_vs_finetune_order_2_results',
       't5_vs_finetune_order_3_results', 't5_vs_finetune_order_4_results',
       't5_vs_finetune_order_5_results', 't5_vs_finetune_order_6_results']
```

```python
row_modes = df_t5[cols_to_compare].mode(axis=1)
row_modes.tail(20)
```

```python
df_t5.tail()
```

```python
df_t5 = pd.concat([df_t5,row_modes], axis=1)

df_t5
```

```python
df_t5.rename(columns={0: 'mode_0', 1: 'mode_1'}, inplace=True)
df_t5
```

```python
df_t5['winner'] = np.where(df_t5['mode_1'].isna(), df_t5['mode_0'], 'none')
df_t5
```

```python
#Build new dataframe to show results of voting
data = {'model': ['t5_fine_tuned', 't5', 'no_winner'],
        'value': [95,63,42]}

df_t5_results = pd.DataFrame(data)
df_t5_results
```

```python
95+63+42
```

```python
fig = plt.figure()

fig.patch.set_facecolor('white')
fig.patch.set_alpha(0)
fig = plt.figure(figsize=(9,4),linewidth=10, facecolor='white')

plt.rcParams['figure.figsize'] = (8,4)
plt.rcParams['figure.dpi'] = 200
plt.rcParams
plt.style.context('seaborn-paper')

colors = sns.color_palette(['#B4B4B4'])
sns.barplot(data=df_t5_results,y=df_t5_results['model'],x=df_t5_results['value'], palette = colors,width=0.4, dodge=False)
plt.xlabel('Number of times the model was chosen as best by LLM (n=200)')
plt.ylabel('')
plt.tick_params(top=False, labeltop=False, bottom=True, labelbottom=True)
#plt.tight_layout()
plt.legend('', frameon=False)
plt.suptitle('Llama3 LLM as Evaluator')
plt.title('T5 vs. T5 fine-tuned', fontsize=10)
#plt.grid(visible=True, axis='x', which='major' )
plt.annotate(text='95', xy=([400,400]), xycoords='axes points', xytext=(452,182), size=15, color='#FF6F27')
plt.annotate(text='63', xy=([500,500]), xycoords='axes points', xytext=(290,106), size=15)
plt.annotate(text='42', xy=([500,500]), xycoords='axes points', xytext=(189,31), size=15)
#plt.annotate(text='18', xy=([500,500]), xycoords='axes points', xytext=(60,31), size=15)
plt.annotate(text='For 42 records, LLM votes did not identify \na winner.  Models tied. ', xy=([217,39]), xycoords='axes points', xytext=(278, 34), 
            arrowprops=dict(arrowstyle='->,widthB=6.2',linewidth=1, connectionstyle='bar,fraction=0'));# textcoords=('axes fraction'), ,
#            arrowprops=dict(arrowstyle='-[,widthB=6.2',color='#FF6F27',linewidth=1, connectionstyle='bar,fraction=0'))
plt.annotate(text='Llama3 preferred T5 fine-tuned\n summaries in 95 of 200 records (47.5%) ', xy=([450,190]), xycoords='axes points', xytext=(285, 145), color='#FF6F27',
            arrowprops=dict(arrowstyle='->,widthB=6.2',linewidth=1, color='#FF6F27', connectionstyle='bar,fraction=0'));# textcoords=('axes fraction'), ,
plt.savefig('data/llm_eval_t5_v_t5_fine_tuned.png', bbox_inches='tight');
```
